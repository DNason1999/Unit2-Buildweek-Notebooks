```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/DNason1999/MtG-MythicChampionship-V-decklists/master/data.csv")
```

```python
def from_list(input_list, columns_in):
  nan = np.nan
  df = pd.DataFrame(data=eval(input_list), columns= columns_in)
  return df

columns0 = ['qty','name','color_identity','colors','mana_cost','cmc','power','toughness','rarity','type_line']

main = df.copy()
main['deck'] = main['deck'].apply(lambda x : from_list(x,columns0))
```

```python
decks = main[['deck']]
decks = decks.reset_index().rename({'index':'rank'}, axis=1)
decks.head()
```

```python
b0 = decks['deck'][0]
b0.set_index('name')
b0.shape
```

```python
 def remove_dupes(input_df):
    output_name = pd.Series()
    output_quant = pd.Series()

    a = input_df.copy(deep=True)

    a0=a[['name','qty']].set_index('name')

    for label, content in a.iterrows():
        if(not output_name.isin([content['name']]).any()):
            output_name= output_name.append(pd.Series(content['name']), ignore_index=True)
            if(type(a0.loc[content['name']]['qty']) == pd.Series):
                output_quant= output_quant.append(pd.Series(sum(a0.loc[content['name']]['qty'].astype(int))), ignore_index=True)
            else:
                output_quant= output_quant.append(pd.Series(a0.loc[content['name']]['qty']), ignore_index=True)
                
    a= a.set_index('name')
    a= a.loc[output_name]
    a= a.reset_index().rename({'index':'name'}, axis=1)
    a.drop_duplicates(subset='name',keep='first', inplace=True)
    a = a.reset_index().drop('index', axis=1)
    a['qty'] = output_quant
    
    return a
```

```python
new_col_0 = []

for x in decks['deck']:
    new_col_0.append(remove_dupes(x))
```

```python
decks['no_dupes'] = new_col_0
```

```python
names = []

for x in decks['no_dupes']:
    for y in x['name']:
        if(not (y in names)):
            names.append(y)
        
```
