```python
df = pd.read_json("https://archive.scryfall.com/json/scryfall-oracle-cards.json")

important_columns = ['id', 'name', 'released_at', 'uri',
             'scryfall_uri', 'image_uris','mana_cost', 'cmc',
             'type_line','colors', 'color_identity','set', 
             'rarity', 'flavor_text', 'artist',
             'power', 'toughness', 'loyalty', 'legalities']

df = df[important_columns]

labels = []

for label,content in df.iterrows():
    if(content['legalities']['standard'] == 'legal'):
        labels.append(label)

main = df.iloc[labels].reset_index().drop(['index','legalities'],axis=1)
```

```python
prices = []
counter = 1

for x in main['uri']:
    error = True
    price = 0.0
    while(error):
        try:
            price = requests.get(x+'?pretty=true').json()['prices']['usd']
            error = False
        except:
            error = True
    prices.append(price)
    time.sleep(0.05)
    print(counter/len(main['uri']), price)
    counter += 1
```

```python
main['price'] = prices
```
